```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Exclude the species column for numerical analysis
iris_numerical = iris.drop(columns=['species'])

# 1. Histograms for distribution of each numerical variable
plt.figure(figsize=(10, 6))
iris_numerical.hist(bins=20, edgecolor='black', figsize=(12, 8))
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)
plt.show()

# 2. Box plots to identify outliers (excluding species)
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_numerical, orient='h', palette='Set2')
plt.title('Box plots of Iris Dataset Features', fontsize=16)
plt.show()

# 3. Heatmap to check for correlations (excluding species)
plt.figure(figsize=(8, 6))
sns.heatmap(iris_numerical.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset', fontsize=16)
plt.show()

# 4. Pair plot to visualize pairwise relationships (including species for coloring)
sns.pairplot(iris, hue='species', palette='Set2', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02, fontsize=16)
plt.show()
```
